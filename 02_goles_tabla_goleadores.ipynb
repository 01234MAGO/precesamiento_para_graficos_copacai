# GOLES 2023

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(database)
query = "SELECT * FROM goles_2023"
df = pd.read_sql_query(query, conn)
conn.close()

df.head(6)
```

# FUNCION PARA CREAR TABLA GOLEADORES

```python
def calcular_goleadores(df):
    df_goleadores = df.groupby(['jugador', 'equipo', 'posicion', 'accion']).size().unstack(fill_value=0)
    df_goleadores['goles'] = df_goleadores.sum(axis=1)
    df_goleadores = df_goleadores.sort_values(by='goles', ascending=False)
    df_goleadores.reset_index(inplace=True)
    return df_goleadores
```

## TABLA GOLEADORES

```python
tabla_goleadores_general = calcular_goleadores(df)
```

## TABLA TOP 8

```python
# eliminar los registros donde la variable "goles" sea 1 y quedan 8 registros
df_goleadores_top8 = tabla_goleadores_general[tabla_goleadores_general['goles'] > 2]
```

## TABLA COMPLETA EN 3 PARTES

```python
# dividr el df que tiene 57 registros en 3 grupos, paras el gruopo 1 tomar los primeros 19 registros, para el segundo tomar los siguientes 19
tabla_goleadores_1 = tabla_goleadores_general[:19]
tabla_goleadores_2 = tabla_goleadores_general[19:38]
tabla_goleadores_3 = tabla_goleadores_general[38:]
```

# DESCARGAS

```python
# tabla de goles
df.to_csv('goles_2023.csv')

# tabla de goleadores
tabla_goleadores_general.to_csv('tabla_goleadores_general.csv')

# tabla de goleadores top 8
df_goleadores_top8.to_csv('tabla_goleadores_top8.csv')

# tabla de goleadores en 3 partes
tabla_goleadores_1.to_csv('tabla_goleadores_1.csv')
tabla_goleadores_2.to_csv('tabla_goleadores_2.csv')
tabla_goleadores_3.to_csv('tabla_goleadores_3.csv')
```
